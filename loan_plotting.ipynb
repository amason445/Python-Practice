```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt  
```

```python
loan = 100000
i = np.array(np.arange(0,.1,0.0001))
i100 = i * 100
n = 8
pv = (1-((1/(1 + i))**n))/i
pmt = loan / pv
monthly_pmt = pmt / 12
```

```python
len(monthly_pmt)
```

```python
plt.title("Monthly Payment as Interest Changes")  
plt.xlabel("Interest Rate")  
plt.ylabel("Monthly Payment")  
plt.plot(i100, monthly_pmt, color ="red")  
plt.show()
```

```python
loan2 = 100000
i2 = 0.0287
n2 = np.arange(1,30)
pv2 = (1-((1/(1 + i2))**n2))/i2
pmt2 = loan2 / pv2
monthly_pmt2 = pmt2 / 12
```

```python
plt.title("Monthly Payment as Term Changes")  
plt.xlabel("Term Years")  
plt.ylabel("Monthly Payment")  
plt.plot(n2, monthly_pmt2, color ="red")  
plt.show()
```
