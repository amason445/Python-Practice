```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# assuming a 10 year, 1.81% coupon bond is bought at premium on issuance for 100.00 (p0)
# and sold at maturity
p0 = 100.00
n = 120
r = 0.00905
Face = 100
yield_range = np.arange(start = 0.0001, stop = 0.12, step = 0.0001)
nCall1 = 20
kArray = Face * ((1/(1 + yield_range))**n)



```

```python
Price1Array = kArray + ((r/yield_range)*(Face - kArray))
```

```python
TotalReturnArray = (((Price1Array - p0) + (nCall1 * (Face * r))) / p0) * 100
```

```python
plt.title("Total Return at Maturity") 
plt.xlabel("Yield %") 
plt.ylabel("Total Return %") 
plt.plot(yield_range * 100,TotalReturnArray) 
plt.show()
```
